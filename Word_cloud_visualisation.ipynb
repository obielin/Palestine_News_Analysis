# Visualization of News Data Using Word Clouds
This notebook aims to analyze and visualize the most frequent words in a cleaned dataset of news articles. By combining headlines and descriptions, we create comprehensive word clouds to identify prevalent themes and subjects. The analysis starts with loading the cleaned news data from Google Drive, followed by the generation of word clouds for both combined text and individual columns (headlines and descriptions). This approach offers a visual exploration of the dataset's content, highlighting the most common words and their significance

```python
from wordcloud import WordCloud
```

### Loading The Dataset
With the environment set up, we load the cleaned news data from a CSV file into a Pandas DataFrame:

```python
#Loading the dataset
file_path = "/content/drive/MyDrive/cleaned_news_data.csv"
df= pd.read_csv(file_path)
df.head(5)
```

### Data Preparation
Before visualization, we prepare the data by combining the 'headline' and 'description' columns into a single text column. This consolidation helps in generating a more comprehensive word cloud:

### Generating Word Clouds
Combined Word Cloud
We generate a word cloud for the combined text of headlines and descriptions to get an overall sense of the most frequent words in the news dataset:

```python
# Combine 'headline' and 'description' columns into one text column
df['text'] = df['headline'] + ' ' + df['description']

# Create a WordCloud object
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(df['text']))

# Display the word cloud using Matplotlib
plt.figure(figsize=(13, 7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("Word Cloud of Most Frequent Words in Headlines and Descriptions")
plt.show()
```

### Separate Word Clouds for Headlines and Descriptions
Additionally, we create separate word clouds for headlines and descriptions to identify the unique themes present in each text type:

### Word Cloud for Headlines:

```python
# Create separate WordCloud objects for 'headline' and 'description'
wordcloud_headline = WordCloud(width=800, height=400, background_color='white').generate(' '.join(df['headline']))


# Display the word clouds using Matplotlib
plt.figure(figsize=(12, 6))
plt.imshow(wordcloud_headline, interpolation='bilinear')
plt.axis("off")
plt.title("Word Cloud for headlines")

plt.tight_layout()
plt.show()
```

### Word Cloud for Descriptions:

```python
# Create separate WordCloud objects for 'headline' and 'description'
wordcloud_description = WordCloud(width=800, height=400, background_color='white').generate(' '.join(df['description']))

# Display the word clouds using Matplotlib
plt.figure(figsize=(13, 7))
plt.imshow(wordcloud_description, interpolation='bilinear')
plt.axis("off")
plt.title("Word Cloud for Description")
```

### Conclusion
These word clouds provide a visual representation of the most frequent words in our news dataset, offering insights into the predominant topics and themes. Such visualizations can help in understanding the focus areas of news coverage and in guiding further analytical or NLP tasks.
